```python
# Example for fetching ticker data for Google from
# alphavantage, by default data is returned in 
# json format, data here is a json file
# Pass your own api key to TimeSeries
from alpha_vantage.timeseries import TimeSeries
ts = TimeSeries('VKJ13X4AAAVHYP2J')
data, meta_data = ts.get_intraday('GOOG')
```

```python
# Specify output_format kwarg as pandas
from pprint import pprint
ts = TimeSeries('VKJ13X4AAAVHYP2J',output_format='pandas')
data, meta_data = ts.get_daily(symbol='GOOG',outputsize='full')
pprint(data.head())
```

```python
plt.style.use('classic')
data['4. close'].plot(title='Google Daily Chart')
```
