```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('netflix.csv')
df.head()
```

```python
df.columns
```

```python
# set_index makes Date as our row index data
# see the difference between 
df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
print(plt.style.available)
```

```python
plt.style.use('classic')
```

```python
df.plot(title='Netflix Closing price',color='purple',figsize=(10,6))
plt.show()
```

```python
returns = df.pct_change()
((1 + returns).cumprod() - 1).plot(title='Netflix Cumu Returns', figsize=(10,6), color='purple')
```

```python
plt.style.use('tableau-colorblind10')
df.pct_change().plot(title='Percent Change',figsize=(10,5),color='blue')
```
